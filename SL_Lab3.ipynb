# **1. Árboles de decisión**
## Definición breve
Método paramétrico no supervisado usado para clasificación y regresión.  La **meta** es crear un modelo que prediga el valor de una variable target aprendiendo reglas de decisión inferidas de las características de la data (scikit-learn.org)

## ¿Cómo funciona el algoritmo?
Divide datos según características importantes

## ¿Para qué casos es bueno este algoritmo?
Datos categóricos o continuos simples

## ¿Para qué casos no es bueno este algoritmo?
Datos ruidos o con muchos outliers

## ¿Cuál es la complejidad computacional de este algoritmo?
O(n log n).

## ¿Qué ventajas tiene este algoritmo sobre los demás?
Fácil interpretación

## ¿Qué desventajas tiene este algoritmo sobre los demás?
Sobreajuste fácil.

## Hiperparámetro (scikit-learn)

*   max_depth: profundidad máxima
*   min_samples_split: mínimos muestras para dividir
*   min_samples_left: mínimos muestras en hoja
*   max_features: máximas características

# **3. K-Nearest Neighbors (KNN)**
## Definición breve
Es un clasificador de aprendizaje supervisado no paramétrico, que utiliza la proximidad para hacer clasificaciones o predicciones sobre la agrupación de un punto de datos individual.
(https://www.ibm.com/mx-es/topics/knn)

## ¿Cómo funciona el algoritmo?
Clasifica por mayoría de vecinos cercanos

## ¿Para qué casos es bueno este algoritmo?
Datos debidamente clasificados o distribuidos

## ¿Para qué casos no es bueno este algoritmo?
Conjunto de datos demasiado grande

## ¿Cuál es la complejidad computacional de este algoritmo?
O(n^2)

## ¿Qué ventajas tiene este algoritmo sobre los demás?
No necesita entrenamiento

## ¿Qué desventajas tiene este algoritmo sobre los demás?
Costoso en tiempo de predicción

## Hiperparámetro (scikit-learn)

*   n_neighbors: número de vecinos
*   algorithm: método de búsqueda
*   leaf_size: tamaño de hojas
*   p: poder de la distancia

# **4.  SVM – Support Vector Machines Lineales**
## Definición breve
SVM funciona correlacionando datos a un espacio de características de grandes dimensiones de forma que los puntos de datos se puedan categorizar, incluso si los datos no se puedan separar linealmente de otro modo.
https://www.ibm.com/docs/es/spss-modeler/saas?topic=models-how-svm-works)

## ¿Cómo funciona el algoritmo?
Encuentra hiperplano de separación óptimo como una frontera de decisión

## ¿Para qué casos es bueno este algoritmo?
Datos linealmente separables

## ¿Para qué casos no es bueno este algoritmo?
Datos con outlier/ruido significativo

## ¿Cuál es la complejidad computacional de este algoritmo?
O(n^2)

## ¿Qué ventajas tiene este algoritmo sobre los demás?
Eficiente en alta dimensionalidad

## ¿Qué desventajas tiene este algoritmo sobre los demás?
No escala bien con grandes datos

## Hiperparámetro (scikit-learn)

*   C: regularización
*   max_iter: iteraciones maximas

# **6. LDA – Análisis de Discriminante Lineal.**

# Definición breve
Es una técnica de clasificación y reducción de dimensionalidad que proyecta los datos en un espacio de menor dimensión, maximizando la separación entre las clases. (https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html)

## ¿Cómo funciona el algoritmo?
Proyección en líneas

## ¿Para qué casos es bueno este algoritmo?
Clases linealmente separables
## ¿Para qué casos no es bueno este algoritmo?
Clases no lineales

## ¿Cuál es la complejidad computacional de este algoritmo?

O(n * d^2)
## ¿Qué ventajas tiene este algoritmo sobre los demás?

Es simple y rápido
## ¿Qué desventajas tiene este algoritmo sobre los demás?
Supone una distribución normal

## Hiperparámetro (scikit-learn)
*   Solver: Método de resolución
*   Shrinkage: Regularización
